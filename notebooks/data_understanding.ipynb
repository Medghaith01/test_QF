The core analytical columns (`Date`, `Mode`, `Gravité`, `Genre`, `Age`, `Milieu`, `Arrondissement`) are fully populated with 0 nulls.

## 4. Duplicate column analysis

The file has multiple columns that appear to be duplicates from bad joins. Let's quantify overlap.

```python
# Coordonnées vs Coordonnées.1 vs Coordonnées.2
print("=== Coordonnées ===")
print(f"Coordonnées nulls:        {df['Coordonnées'].isnull().sum()}")
print(f"Coordonnées.1 nulls:       {df['Coordonnées.1'].isnull().sum()}")
print(f"Coordonnées.2 nulls:       {df['Coordonnées.2'].isnull().sum()}")
print(f"Coordonnées.1 == .2 (all): {(df['Coordonnées.1'] == df['Coordonnées.2']).all()}")
mask = df['Coordonnées'].notna()
print(f"Where non-null, .1 == .2:  {(df.loc[mask, 'Coordonnées'] == df.loc[mask, 'Coordonnées.1']).all()}")
print(f"Rows where . is null but .1 has data: {df['Coordonnées'].isnull().sum() - df['Coordonnées.1'].isnull().sum()}")
```

**Observations:** `Coordonnées.1` and `Coordonnées.2` are 100% identical and fully populated, while `Coordonnées` has 10,235 nulls. Where `Coordonnées` is non-null, it matches `.1` exactly. This confirms `.1` and `.2` are redundant duplicates — `Coordonnées.1` should be used as the canonical coordinate column since it has no missing values.

```python
# Nom arrondissement vs .1 vs .2
print("=== Nom arrondissement ===")
print(f"Nom arrondissement nulls:        {df['Nom arrondissement'].isnull().sum()}")
print(f"Nom arrondissement.1 nulls:       {df['Nom arrondissement.1'].isnull().sum()}")
print(f"Nom arrondissement.2 nulls:       {df['Nom arrondissement.2'].isnull().sum()}")
print(f".1 == .2 (all):                  {(df['Nom arrondissement.1'] == df['Nom arrondissement.2']).all()}")
```

**Observations:** Same pattern as coordinates: `Nom arrondissement.1` and `.2` are identical and complete, while the original `Nom arrondissement` is missing 10,235 values. Use `Nom arrondissement.1` as the canonical column.

```python
# Arrondissement vs .1 vs .2
print("=== Arrondissement ===")
print(f"Arrondissement == Arrondissement.1: {(df['Arrondissement'] == df['Arrondissement.1']).all()}")
print(f"Arrondissement == Arrondissement.2: {(df['Arrondissement'] == df['Arrondissement.2']).all()}")
print(f"Arrondissement.2 dtype: {df['Arrondissement.2'].dtype}")
print(f"Arrondissement.2 unique values: {sorted(df['Arrondissement.2'].dropna().unique())[:10]}...")
suspect = df[df['Arrondissement'].astype(str) != df['Arrondissement.2'].astype(str)]
print(f"Rows where .2 doesn't match: {len(suspect)}")
```

**Observations:** `Arrondissement` and `Arrondissement.1` are identical (both `int64`). However, `Arrondissement.2` is a `float64` with different values — it includes a mix of department codes (75) and full arrondissement codes (75101–75120), causing 33,551 mismatches. `Arrondissement.2` appears to come from a different join and should be dropped; use `Arrondissement` (or `.1`) as the canonical column.

```python
# Adresse vs Adresse 1 vs Adresse 2
print("=== Adresse ===")
print(f"Adresse nulls:     {df['Adresse'].isnull().sum()}")
print(f"Adresse 1 nulls:   {df['Adresse 1'].isnull().sum()}")
print(f"Adresse 2 nulls:   {df['Adresse 2'].isnull().sum()}")
mask = df['Adresse'].notna() & df['Adresse 1'].notna()
print(f"Adresse == Adresse 1 (where both non-null): {(df.loc[mask, 'Adresse'] == df.loc[mask, 'Adresse 1']).sum()}/{mask.sum()}")
print(f"\nThese are DIFFERENT address components, not duplicates.")
```

**Observations:** The three address columns are not duplicates — they represent different components. `Adresse` is the primary address (almost complete, only 3 nulls), while `Adresse 1` and `Adresse 2` are cross-street / secondary address fields with high null rates (69.5% and 80.5%). Among the 10,232 rows where both `Adresse` and `Adresse 1` are non-null, only 43% match, confirming they hold different street names (likely the two streets at an intersection).

```python
# arrondgeo check
print(f"arrondgeo nulls: {df['arrondgeo'].isnull().sum()}")
print(f"arrondgeo sample type: {type(df['arrondgeo'].iloc[0])}")
print(f"arrondgeo contains GeoJSON MultiPolygon polygons for each arrondissement.")
print(f"Always populated — pure spatial metadata, not a textual duplicate.")
```

**Observations:** `arrondgeo` contains GeoJSON `MultiPolygon` geometries defining the boundary of each Paris arrondissement. It is always populated and is purely spatial metadata used for mapping — not a duplicate of the textual arrondissement name columns.

## 6. Injury indicator consistency check

```python
# Check that each row has exactly one injury flag set
flags = df[["Blessés Légers", "Blessés hospitalisés", "Tué"]]
flag_sum = flags.notna().sum(axis=1)
print("Rows with 0 flags set:", (flag_sum == 0).sum())
print("Rows with 1 flag set:", (flag_sum == 1).sum())
print("Rows with 2+ flags set:", (flag_sum >= 2).sum())

# Verify they match Gravité
df["_bl"] = df["Blessés Légers"].notna()
df["_bh"] = df["Blessés hospitalisés"].notna()
df["_tu"] = df["Tué"].notna()
df["_match"] = (
    ((df["Gravité"] == "Blessé léger") & df["_bl"]) |
    ((df["Gravité"] == "Blessé hospitalisé") & df["_bh"]) |
    ((df["Gravité"] == "Tué") & df["_tu"])
)
print(f"\nGravité matches its indicator column: {df['_match'].all()}")
df.drop(columns=["_bl", "_bh", "_tu", "_match"], inplace=True)
```

**Observations:** Each row has at least one injury flag set, and 33,519 out of 33,551 rows (99.9%) have exactly one flag — confirming the three injury indicator columns are mutually exclusive. The 32 rows with 2+ flags are rare anomalies (likely data entry errors). The `Gravité` column is perfectly consistent with its corresponding indicator column. These three flags can serve as binary features or be used to validate the `Gravité` field.
